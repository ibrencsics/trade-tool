```python
import pandas as pd
```

# 1: Dividend champions
https://moneyzine.com/investments/dividend-champions/

```python
champ = pd.read_excel("U.S.DividendChampions-LIVE.xlsx", sheet_name="Champions", skiprows=2, index_col="Symbol")
```

```python
champ = champ[["Company", "Sector", "Industry", "No Years", "Price", "P/E", "Div Yield", "5Y Avg Yield", "Current Div", "Previous Div", "DGR 1Y", "DGR 3Y", "DGR 5Y", "DGR 10Y"]]
champ["MR%"] = (champ["Current Div"] - champ["Previous Div"]) / champ["Previous Div"] * 100
```

# 4: Div Yield between S&P x1.5 and x5
https://www.multpl.com/s-p-500-dividend-yield

Also bigger than inflation
https://www.multpl.com/inflation

```python
snp = 1.56
inflation = 3.18
snp_range_lower = champ["Div Yield"] > snp * 1.5
snp_range_upper = champ["Div Yield"] < snp * 5
bigger_than_inflation = champ["Div Yield"] > inflation
champ = champ[snp_range_lower & snp_range_upper & bigger_than_inflation]
```

```python
len(champ)
```

```python
champ.sort_values(by = "Div Yield", ascending = False).head(1)
```

# DGR growing

```python
div_growth_1_3 = champ["DGR 1Y"] > champ["DGR 3Y"]
div_growth_3_5 = champ["DGR 3Y"] > champ["DGR 5Y"]
div_growth_5_10 = champ["DGR 5Y"] > champ["DGR 10Y"]
champ = champ[div_growth_1_3 & div_growth_3_5 & div_growth_5_10]
```
